**Q4.** Optimization is at the core of data science and statistics: Picking the best fit or estimate according to some desirable criteria (e.g. Maximum Likelihood Estimation). In this question, you're going to write a function that finds the highest possible value for some function, and returns the value of the function and the maximizing value.

1. Write a function that creates a grid from some value $a$ to a second value $b$ with $N$ steps, computes the value of a function $f$ on that grid, and then finds the maximizers of the function. Have it return a dictionary with the maximum value and the maximizers themselves.
2. Use your function to maximize $f(x) = 100 - 2x^2 + 3x$ for values of $a=-1$, $b=1$, and for values of $N=3$, $N=10$, $N=100$, $N=1000$ and $N=5000$.
3. The true maximizer of this function is $.75$ and the true maximum value is $101.125$. Is that what you got in the previous step? Why not? How does the quality of the maximization depend on $N$? How does your computed answer change as $N$ gets larger?

  My results gave me a maximizer of 0.75, and the 101.125 when rounded, so it matches the true maximizer and true maximum value. As the N value gets larger, the maximizer and maximum values will begin to converge at a certain point since the points get closer. Opposed to when the N value is smaller, the maximizer and maximum may not be an accurate representation since the points are not as densely compact together.

```python
import numpy as np
import matplotlib.pyplot as plt

# Used ChatGPT as starter code
def maximize_function(f, a, b, N):
  x_val = np.linspace(a, b, N)
  f_val = f(x_val)
  max_val = np.max(f_val)
  max_x_val = x_val[f_val == max_val]
  return {'max_val': max_val, 'maximizers': max_x_val}


def f(x):
  return 100 - 2 * x**2 + 3 * x

a = -1
b = 1

n_values = [3, 10, 100, 1000, 5000]
results = {}

for N in n_values:
  result = maximize_function(f, a, b, N)
  results[N] = result
  print(f"N = {N}: Max value = {result['max_val']}, Maximizers = {result['maximizers']}")

x_plot = np.linspace(a, b, 1000)
y_plot = f(x_plot)

plt.figure(figsize=(8, 5))
plt.plot(x_plot, y_plot, label='f(x) = 100 - 2x^2 + 3x')
plt.scatter(results[5000]['maximizers'], results[5000]['max_val'], color='red', label='Maximizers (N=5000)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function Maximizer')
plt.legend()
plt.grid(True)
plt.show()

```
